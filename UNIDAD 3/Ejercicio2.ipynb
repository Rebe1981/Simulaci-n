```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de llegadas)
lambda_llegadas = 1 / 15  # En promedio, 1 llegada cada 15 minutos

# Función para generar un tiempo entre llegadas
def generar_tiempo_entre_llegadas(lambda_llegadas):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo = -1 / lambda_llegadas * np.log(1 - u)  # Transformada inversa
    return tiempo

# Generar tiempos entre llegadas hasta completar las 12 horas de operación
tiempos_entre_llegadas = []
tiempo_acumulado = 0
minutos_operacion = 12 * 60  # Convertir 12 horas a minutos

while tiempo_acumulado < minutos_operacion:
    tiempo = generar_tiempo_entre_llegadas(lambda_llegadas)
    tiempos_entre_llegadas.append(tiempo)
    tiempo_acumulado += tiempo

# Mostrar resultados
print("Tiempos entre llegadas generados:", tiempos_entre_llegadas[:5])  # Mostramos solo los primeros 5
print(f"Total de vehículos simulados: {len(tiempos_entre_llegadas)}")

# Graficar los tiempos generados
plt.hist(tiempos_entre_llegadas, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas de Vehículos (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```

El uso de variables aleatorias para simular tiempos de servicio, atención e inactividad ofrece varios beneficios:

Modelado de procesos impredecibles: Permite simular eventos aleatorios en situaciones reales, como llegadas de vehículos o llamadas a un centro de atención.

Optimización de recursos: Ayuda a determinar la cantidad adecuada de recursos necesarios para evitar sobrecarga y mejorar la eficiencia operativa.

Mejora de tiempos de respuesta: Permite identificar cuellos de botella y optimizar los tiempos de espera o inactividad, mejorando la experiencia del cliente.

Toma de decisiones informadas: Facilita la toma de decisiones sobre capacidad y recursos, basándose en simulaciones de escenarios posibles.

Versatilidad: Estas simulaciones son aplicables en diversos sectores como logística, telecomunicaciones y atención al cliente.

Mejora continua: Las simulaciones repetitivas permiten ajustar modelos y mejorar continuamente los procesos.
